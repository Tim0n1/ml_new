```python
def func(X=None, Y=None):
    if Y is None:
        return X**2
    elif X is None:
        return np.exp(np.abs(Y)) + 1/Y
    else:
        return X**2 + np.exp(np.abs(Y)) + 1/Y

def derivative(func, x, h=1e-5, parameter=None):
    """
    Compute the derivative of a function at a given point using the central difference method.

    Parameters:
        func (callable): The function to compute the derivative of.
        x (float): The point at which to compute the derivative.
        h (float): The step size for the central difference method. Default is 1e-5.

    Returns:
        float: The approximate derivative of the function at the given point.
"""
    if parameter == 'y':
        # Compute f(x + h)
        f_plus_h = func(Y=(x + h))
        
        # Compute f(x - h)
        f_minus_h = func(Y=(x - h))
        
        # Compute the central difference approximation of the derivative
        derivative_approx = (f_plus_h - f_minus_h) / (2 * h)
        return derivative_approx
    else:
         # Compute f(x + h)
        f_plus_h = func(X=(x + h))
        
        # Compute f(x - h)
        f_minus_h = func(X=(x - h))
        
        # Compute the central difference approximation of the derivative
        derivative_approx = (f_plus_h - f_minus_h) / (2 * h)
        return derivative_approx
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def f(x, y):
    return np.sin(x) * np.cos(y)

# Define the gradient of the function
def gradient_f(x, y):
    df_dx = np.cos(x) * np.cos(y)
    df_dy = -np.sin(x) * np.sin(y)
    return df_dx, df_dy

# Generate data
x = np.linspace(-2*np.pi, 2*np.pi, 50)
y = np.linspace(-2*np.pi, 2*np.pi, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Compute gradient at each point
df_dx, df_dy = gradient_f(X, Y)

# Plot figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Plot gradient vectors
ax.quiver(X[:10], Y[:10], Z[:10], df_dx[:10], df_dy[:10], np.zeros_like(Z[:10]), length=0.3, color='red', normalize=True)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot with Gradient Vectors')

plt.show()
```
