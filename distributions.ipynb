```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde

# Generate sample data
np.random.seed(42)
x = np.random.standard_normal(256)
y = np.random.standard_normal(256)

# Perform a 2D KDE on the data
kde = gaussian_kde(np.vstack([x, y]))

# Create a grid of points
x_min, x_max = x.min() - 1, x.max() + 1
y_min, y_max = y.min() - 1, y.max() + 1
X, Y = np.mgrid[x_min:x_max:100j, y_min:y_max:100j]
positions = np.vstack([X.ravel(), Y.ravel()])
Z = np.reshape(kde(positions).T, X.shape)

# Integrate over x to get the marginal distribution of y
Pr_y = np.trapz(Z, x=X[:, 0], axis=0)

# Plot the joint KDE
plt.figure(figsize=(12, 8))
sns.kdeplot(x=x, y=y, fill=True, color="r")
plt.grid(True)
plt.title('Joint KDE of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Plot the marginal distribution of y
plt.figure(figsize=(12, 8))
plt.plot(Y[0, :], Pr_y, color='b')
plt.fill_between(Y[0, :], Pr_y, color='b', alpha=0.3)
plt.grid(True)
plt.title('Marginal Distribution of y')
plt.xlabel('y')
plt.ylabel('Density')
plt.show()
```
